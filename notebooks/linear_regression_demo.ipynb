```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)  # For reproducibility
X = np.linspace(0, 10, 50)  # 50 points between 0 and 10
noise = np.random.normal(0, 1, 50)  # Add some noise
y = 2 * X + 1 + noise  # True relationship: y = 2X + 1 + noise

# Reshape X to (m, 1) and add a column of ones for the intercept term
X_b = np.c_[np.ones((50, 1)), X.reshape(-1, 1)]  # Shape: (50, 2)

# Initialize theta (intercept and slope)
theta = np.random.randn(2, 1)  # Random initialization

# Gradient Descent Parameters
alpha = 0.01  # Learning rate
iterations = 20  # Number of iterations

# Gradient Descent
for _ in range(iterations):
    gradients = (2 / 50) * X_b.T @ (X_b @ theta - y.reshape(-1, 1))  # Gradient calculation
    theta -= alpha * gradients  # Update theta

# Predict using the final theta
y_pred = X_b @ theta  # Matrix multiplication for predictions

# Plot the results
plt.scatter(X, y, label='Data Points')  # Scatter plot of the data
plt.plot(X, y_pred, color='red', label='Regression Line: $y = \\theta_0 + \\theta_1 X$')  # Regression line

# Add labels, title, and legend
plt.xlabel('X')
plt.ylabel('y')
plt.title('Synthetic Data for Linear Regression')
plt.legend()
plt.show()

# Print the final theta values
print("Final theta (intercept and slope):", theta.flatten())
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(0,10,100)
noice = np.random.normal(0,1,100)
y= 2 * X + 1 + noice 
theta = np.random.randn(2,1)
X_b = np.c_[np.ones((100,1)),X.reshape(-1,1)]
alpha = 0.01
iterations = 200
for _ in range(iterations) :
    gradients = (2/100)  * X_b.T @ (X_b @ theta - y.reshape(-1,1))
    theta -= gradients * alpha
y_pred = X_b @ theta
plt.scatter(X,y)
plt.plot(X,y_pred ,color="red")
plt.show()
```
